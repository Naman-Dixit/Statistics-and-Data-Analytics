```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import io
from google.colab import files
```

```python
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['salary (5).csv']))
print(df.head(10))
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['salary'], bins=20)
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.title('Salary Distribution Histogram')
plt.show()
```

```python
df.shape
```

```python
plt.figure(figsize=(8,5))
plt.boxplot(df['salary'])
plt.ylabel('Salary')
plt.title("Boxplot of Salary Distribution")
plt.show()
```

```python
plt.figure()
sns.kdeplot(df['salary'], fill=True)
plt.xlable("Salary")
plt.ylabel("Density")
plt.title("Kernel Density Estimation of Salary Distribution")
plt.show()
```
